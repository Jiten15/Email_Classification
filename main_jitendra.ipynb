```python
X_train, X_test = train_test_split(df, test_size=0.2, random_state=111)
```

```python
from sklearn.utils import class_weight
```

```python
class_weights = list( class_weight.compute_class_weight(class_weight = 'balanced',classes = np.unique(df['CATEGORY']), y = df['CATEGORY']) )
                                             
```

```python
df['CATEGORY'].value_counts()
```

```python
class_weights.sort()
```

```python
class_weights
```

```python
weights={}
```

```python
for index, weight in enumerate(class_weights) :
  weights[index]=weight
```

```python
weights
```

```python
dataset_train = tf.data.Dataset.from_tensor_slices((X_train['TITLE'].values, X_train['CATEGORY'].values))
dataset_test = tf.data.Dataset.from_tensor_slices((X_test['TITLE'].values, X_test['CATEGORY'].values))
```

```python
for text, target in dataset_train.take(5):
  print ('Complaint: {}, Target: {}'.format(text, target))
```

```python
for text, target in dataset_test.take(5):
  print ('Complaint: {}, Target: {}'.format(text, target))
```

```python
table = tf.lookup.StaticHashTable(
    initializer=tf.lookup.KeyValueTensorInitializer(
        keys=tf.constant(['Business', 'Personal', 'Personmal_professional', 'logistic_arrangements', 'Employment_arrangement', 'Documen-editing']),
        values=tf.constant([0, 1, 2, 3, 4, 5]),
    ),
    default_value=tf.constant(-1),
    name="target_encoding"
)

@tf.function
def target(x):
  return table.lookup(x)
```

```python
def show_batch(dataset, size=5):
  for batch, label in dataset.take(size):
      print(batch.numpy())
      print(target(label).numpy())
```

```python
show_batch(dataset_test,6)
```

```python
def fetch(text, labels):
  return text, tf.one_hot(target(labels),6)
```

```python
train_data_f=dataset_train.map(fetch)
test_data_f=dataset_test.map(fetch)
```

```python
next(iter(train_data_f))
```

```python
train_data, train_labels = next(iter(train_data_f.batch(5)))
train_data, train_labels
```

```python
embedding = "https://tfhub.dev/google/tf2-preview/nnlm-en-dim128/1"
hub_layer = hub.KerasLayer(embedding, output_shape=[128], input_shape=[], 
                           dtype=tf.string, trainable=True)
hub_layer(train_data[:1])
```

```python
model = tf.keras.Sequential()
model.add(hub_layer)
for units in [128, 128, 64 , 32]:
  model.add(tf.keras.layers.Dense(units, activation='relu'))
  model.add(tf.keras.layers.Dropout(0.3))
model.add(tf.keras.layers.Dense(6, activation='softmax'))

model.summary()
```

```python
model.compile(optimizer='adam',
              loss=tf.keras.losses.CategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
train_data_f=train_data_f.shuffle(70000).batch(64)
test_data_f=test_data_f.batch(64)
```

```python
history = model.fit(train_data_f,
                    epochs=12,
                    validation_data=test_data_f,
                    verbose=1,
                    class_weight=weights)
```

```python
len(list(dataset_test))
```

```python
results = model.evaluate(dataset_test.map(fetch).batch(1663), verbose=2)

print(results)
```
