```python

import pandas as pd
import string
from nltk.stem.porter import PorterStemmer
import nltk
from nltk.stem import WordNetLemmatizer
import re
#whitespace tokenizer
from nltk.tokenize import WhitespaceTokenizer
from textblob import Word

nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('stopwords')
#defining the object for stemming
porter_stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()
```

```python
stopwords = nltk.corpus.stopwords.words('english')
```

```python
def remove_urls(text):
      url_pattern = re.compile(r'https?://\S+|www\.\S+')
      return url_pattern.sub(r'', text)

def remove_html(text):
      html_pattern = re.compile('<.*?>')
      return html_pattern.sub(r'', text)

#defining the function to remove punctuation
def remove_punctuation(text):
    punctuationfree="".join([i for i in text if i not in string.punctuation])
    return punctuationfree  

def tokenization(text):
    tk = WhitespaceTokenizer()
    return tk.tokenize(text)


#defining the function to remove stopwords from tokenized text
def remove_stopwords(text):
    output= [i for i in text if i not in stopwords]
    return output



#defining a function for stemming
def stemming(text):
    stem_text = [porter_stemmer.stem(word) for word in text]
    return stem_text


#spelling correction
def spelling_check(dat):
    new_dat=[]
    for word in dat:
      word = Word(word)
      result = word.spellcheck()
      new_dat.append(result[0][0])
    return  new_dat


def preprocess_data(dat):
  dat = dat.lower()
  dat = remove_urls(dat) 
  dat = remove_html(dat)
  dat = ''.join(c for c in dat if not c.isdigit())
  dat = remove_punctuation(dat)

  from nltk.corpus import stopwords
  ", ".join(stopwords.words('english'))

  #remove stopword
  #defining function for tokenization
  
  #applying function to the column for making tokens in both Training and Testing data
  dat = tokenization(dat)

  #Stop words present in the library
  stopwords = nltk.corpus.stopwords.words('english')
  
  
  
  #applying the function for removal of stopwords
  dat = remove_stopwords(dat)

  #spelling correction
#   dat = spelling_check(dat)
  
  # applying function for stemming
  dat = stemming(dat)

  #Lemmitization


  lemm_words=[]

  for word in dat:
    
    proper_word = lemmatizer.lemmatize(word)
    lemm_words.append(proper_word)

  dat = lemm_words

  # print(dat)    
  #making text out of words


  return " ".join(dat)

```
